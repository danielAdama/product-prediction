## Outlier Detection using the IQR (Inter Quantile Range)

```python
def detect_outliers(dataframe, col):
    
    """
    Function to detect the presence of outliers in a column.
    
    Args:
        dataframe (Dataframe) : dataframe of the data
        col (string) : column name of the data
        
    Returns:
        float : detected outliers
    """
    
    # 1st quantile 2%
    Q1 = dataframe[col].quantile(0.02)
    # 3rd quantile 98%
    Q3 = dataframe[col].quantile(0.98)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    
    print(f"Q1 is {Q1} while Q3 is {Q3}")
    print(f"IQR is {IQR}")
    print(f"The Lower limit is {lower_limit} while the Upper limit is {upper_limit}")
    # Return the detected Outliers
    return data.loc[(dataframe[col] < lower_limit) | (dataframe[col] > upper_limit)]

def treat_right_skewed(dataframe, col):
    """
    Function to treat the right skew present in a column.
    
    Args:
        dataframe (DataFrame) : dataframe of the data
        col (string) : column name
        
    Returns:
        float : with no outliers
    """
    q = dataframe[col].quantile(0.98)
    return dataframe.loc[dataframe[col] < q]

def treat_left_skewed(dataframe, col):
    """
    Function to treat the left skew present in a column.
    
    Args:
        dataframe (DataFrame) : dataframe of the data
        col (string) : column name
        
    Returns:
        float : with no outliers
    """
    q = dataframe[col].quantile(0.02)
    return dataframe.loc[dataframe[col] > q]
```

# 7. Evaluation Metrics
It is not necessary that the model has good metrics
- Answer: Good model metrics matters because we want to achieve high Recall score which produce better model predictions.

Since this is a Binary Classification problem, we can use the 2 following popular metrics.
1. Recall
2. Area under the Receiver Operating Characteristic Curve

Now, we're looking at the recall value, because a product that was mistakenly listed as a no product is just as terrible as a product that was not recognized as a no product and the store failed to take the necessary steps to resupply it.

The ROC AUC is the area under the curve when plotting the (normalized) true positive rate (y-axis) and the false positive rate (x-axis).

Our main metric here would be Recall values, while AUC ROC Score would take care of how well predicted probabilites are able to differentiate between the 2 classes.

```python
cm = confusion_matrix(y_test, y_pred)

def confusion_matrix_plot(cm, normalized= True, cmap = 'bone'):
    
    """Function to visualize how well the model performed on the test set."""
    
    norm_cm = cm
    if normalized:
        plt.figure(figsize=(6,4))
        norm_cm = (cm.astype('float')/ cm.sum(axis= 1)[:, np.newaxis])
        return sns.heatmap(norm_cm, annot= cm, fmt='g', 
                           xticklabels= ['Predicted: No', 'Predicted: Yes'], 
                           yticklabels=['Actual: No', 'Actual: Yes'])

print(f"ROC AUC score of Logistic Regression: {roc_auc_score(y_test, prob_pred):.4f}")
print(f"Recall score of Logistic Regression: {recall_score(y_test, y_pred):.4f}")
confusion_matrix_plot(cm)
plt.show()
```

The ROC AUC score is high which is a good thing but the Recall score is low, we will use cross validation to improve the Recall score.

In the confusion matrix, out of 5599 samples of no products the model predicted 5576 correctly and 23 wrong predictions while for the availability of products the model predicted 480 correctly and 998 wrongly for 1478 samples.

# 7.3 Model Evaluation using Cross Validation

```python
def cross_validation_score(ml_model, thres = 0.5, random_st=42, cols = data_cleaned.columns):
    
    """
    Function to calculate the k fold cross validation stratified on the basis of target
    and prints the ROC, Recall and Precision Scores.
    
    Args:
        ml_model (numpy array) : ml_model for predictions
        thres (float) : threshold for the probabilities of the model predictions
        random_st (int) : random_st is the random state for Kfold
        cols (string) : cols are the column names
    
    Returns:
        cv_scores (float) : cross validation scores
    """
    
    i= 1
    x1 = data_cleaned[cols]
    cv_scores = []
    
    sKf = StratifiedKFold(n_splits= 5, shuffle= True, random_state= random_st)
    
    for train_idx, test_idx in sKf.split(x1, y):
        print(f"\n{i} of KFold {sKf.n_splits}")
        
        # split the data
        xtr, xval = x1.iloc[train_idx],x1.iloc[test_idx]
        ytr, yval = y.iloc[train_idx], y.iloc[test_idx]
        
        model = ml_model
        model.fit(xtr, ytr)
        pred_probs = model.predict_proba(xval)
        pp = []
        
        # Use threshold to define the classes based on probability values
        for j in pred_probs[:,1]:
            if j > thres:
                pp.append(1)
            else:
                pp.append(0)
                
        # Calculate scores for each fold
        pred_val = pp
        roc_score = roc_auc_score(yval, pred_probs[:,1])
        recall = recall_score(yval, pred_val)
        precision = precision_score(yval, pred_val)
        msg = ""
        msg += f"ROC AUC Score: {roc_score:.4f}, Recall Score: {recall:.4f}, Precision Score: {precision:.4f}"
        print(f"{msg}")
        cv_scores.append(roc_score)
        i+=1
        
    return cv_scores
```
